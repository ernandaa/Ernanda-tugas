```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# ---------- 1. DEFINISI VARIABEL ----------
suhu = ctrl.Antecedent(np.arange(35, 41, 0.1), 'suhu')
trombosit = ctrl.Antecedent(np.arange(50, 500, 1), 'trombosit')
leukosit = ctrl.Antecedent(np.arange(2, 15, 0.1), 'leukosit')

risiko = ctrl.Consequent(np.arange(0, 10, 0.1), 'risiko')

# ---------- 2. MEMBERSHIP FUNCTION ----------

# Suhu tubuh (°C)
suhu['normal'] = fuzz.trimf(suhu.universe, [35, 36.5, 37.2])
suhu['demam'] = fuzz.trimf(suhu.universe, [37.0, 38.0, 38.5])
suhu['tinggi'] = fuzz.trimf(suhu.universe, [38.0, 39.5, 41])

# Trombosit (10^9/L)
trombosit['rendah'] = fuzz.trimf(trombosit.universe, [50, 100, 150])
trombosit['normal'] = fuzz.trimf(trombosit.universe, [150, 250, 450])

# Leukosit (10^9/L)
leukosit['rendah'] = fuzz.trimf(leukosit.universe, [2, 3.5, 4.5])
leukosit['normal'] = fuzz.trimf(leukosit.universe, [4, 7, 11])

# Output risiko
risiko['rendah'] = fuzz.trimf(risiko.universe, [0, 2, 4])
risiko['sedang'] = fuzz.trimf(risiko.universe, [3, 5, 7])
risiko['tinggi'] = fuzz.trimf(risiko.universe, [6, 8, 10])

# ---------- 3. RULE BASE ----------

rule1 = ctrl.Rule(suhu['tinggi'] & trombosit['rendah'] & leukosit['rendah'], risiko['tinggi'])
rule2 = ctrl.Rule(suhu['demam'] & trombosit['rendah'], risiko['tinggi'])
rule3 = ctrl.Rule(suhu['demam'] & trombosit['normal'], risiko['sedang'])
rule4 = ctrl.Rule(suhu['normal'] & trombosit['normal'] & leukosit['normal'], risiko['rendah'])
rule5 = ctrl.Rule(suhu['tinggi'] & leukosit['normal'], risiko['sedang'])

# ---------- 4. SISTEM KENDALI ----------
risiko_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5])
risiko_simulasi = ctrl.ControlSystemSimulation(risiko_ctrl)

# ---------- 5. INPUT DATA PASIEN ----------
# contoh soal:
# Suhu 38.5°C, Trombosit 120, Leukosit 3.8
risiko_simulasi.input['suhu'] = 38.5
risiko_simulasi.input['trombosit'] = 120
risiko_simulasi.input['leukosit'] = 3.8

# ---------- 6. HITUNG ----------
risiko_simulasi.compute()

print("HASIL DIAGNOSA RISIKO DBD : ", risiko_simulasi.output['risiko'])
```
